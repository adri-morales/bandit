# Results Analysis

### Imports

```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
from glob import glob
```

## Load data

```python
df_list = []
for history in glob('../data/results/*.csv'):
    df = pd.read_csv(history, index_col=0)
    df_list.append(df)
```

```python
df = pd.concat(df_list)
```

```python
df = pd.DataFrame(df.groupby(['step', 'agent']).reward.mean()).reset_index()
```

```python
sns.lineplot(data=df, x='step', y='reward', hue='agent')
plt.show()
```
